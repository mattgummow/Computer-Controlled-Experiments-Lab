```python
from scipy.optimize import curve_fit
from numpy import inf
```

```python
# Exprimenting with shockley fit 
# Equation simply performs shockley diode equation

def shockley_fit(voltage, current_saturated, ideality_factor, thermal_voltage):
    return current_saturated * (np.exp((voltage)/(ideality_factor*thermal_voltage))-1)
```

```python
# Attempting to curve fit for shockley equation

shock_params, shock_cov = curve_fit(shockley_fit, blue_LED['Vol. across LED (V)'], blue_LED['Current (A)'], p0 = (1e-12, 1, 26e-3), bounds = ([0,1,0], [inf,3, inf]))
cursaturation_shock, ideality_shock, thermvol_shock = shock_params
cursaturation_error = np.sqrt(shock_cov[0,0])
ideality_error = np.sqrt(shock_cov[1,1])
thermvol_error = np.sqrt(shock_cov[2,2])
shockley_currents = shockley_fit(blue_LED['Vol. across LED (V)'].values, cursaturation_shock, ideality_shock, thermvol_shock)

# Plotting on graph

plt.figure(figsize = (10,6))

plt.plot(blue_LED['Vol. across LED (V)'].values, blue_LED['Current (A)'].values, color = 'blue', label = 'LED measurements')
plt.plot(blue_LED['Vol. across LED (V)'].values, shockley_currents, color = 'red', label = 'Shockley fitted current', marker = '.')


legend = plt.legend(loc= 'upper left')
plt.xlabel('Voltage (V)')
plt.ylabel('Current (mA)')
plt.title('Blue LED measurements compared with Shockley fitted current')
plt.show()

print(' Standard error of reverse saturation current is: ' + str(cursaturation_error))
print(' Standard error of ideality factor is: ' + str(ideality_error))
print(' Standard deviation of thermal voltage is: ' + str(thermvol_error))
```
